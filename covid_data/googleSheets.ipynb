```python
sheet_url = 'https://docs.google.com/spreadsheets/d/1hKYHSMNcGvOgVJik4Sg2dSMGvala9qox7RJFaKYBUzw/edit#gid=0'
```

```python
csv_export_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
```

```python
df_states = pd.read_csv(csv_export_url)
```

```python
df_states.drop(df_states.columns[7:12], axis=1, inplace=True)
```

```python
df_states.columns = ['Date', 'State', 'deathsRunning', 'hospRunning', 'icuRunning', 'casesRunning', 
                    'testsRunning', 'Deaths', 'Hosp', 'ICU', 'Cases', 'Tests']
```

```python
df_states = df_states.iloc[1:]
```

```python
df_states.head()
```

```python
df_states.replace(',','', regex=True, inplace=True)
```

```python
df_states['Cases'] = pd.to_numeric(df_states['Cases'])
```

```python
df_states.info()
```

```python
df_states['7sum_posCases'] = df_states.groupby('State')['Cases'].rolling(7).sum().shift(-6).reset_index(0,drop=True)
df_states['14sum_posCases'] = df_states.groupby('State')['Cases'].rolling(14).sum().shift(-13).reset_index(0,drop=True)
df_states['21sum_posCases'] = df_states.groupby('State')['Cases'].rolling(21).sum().shift(-20).reset_index(0,drop=True)
df_states['7sum_totTests'] = df_states.groupby('State')['Tests'].rolling(7).sum().shift(-6).reset_index(0,drop=True)
df_states['14sum_totTests'] = df_states.groupby('State')['Tests'].rolling(14).sum().shift(-13).reset_index(0,drop=True)
df_states['21sum_totTests'] = df_states.groupby('State')['Tests'].rolling(21).sum().shift(-20).reset_index(0,drop=True)
df_states['7ma_positiveTestRate'] = df_states['7sum_posCases'] / df_states['7sum_totTests']
df_states['14ma_positiveTestRate'] = df_states['14sum_posCases'] / df_states['14sum_totTests']
df_states['21ma_positiveTestRate'] = df_states['21sum_posCases'] / df_states['21sum_totTests']
df_states['7sum_totHosp'] = df_states.groupby('State')['Hosp'].rolling(7).sum().shift(-6).reset_index(0,drop=True)
df_states['14sum_totHosp'] = df_states.groupby('State')['Hosp'].rolling(14).sum().shift(-13).reset_index(0,drop=True)
df_states['21sum_totHosp'] = df_states.groupby('State')['Hosp'].rolling(21).sum().shift(-20).reset_index(0,drop=True)
df_states['7sum_death'] = df_states.groupby('State')['Deaths'].rolling(7).sum().shift(-6).reset_index(0,drop=True)
df_states['14sum_death'] = df_states.groupby('State')['Deaths'].rolling(14).sum().shift(-13).reset_index(0,drop=True)
df_states['21sum_death'] = df_states.groupby('State')['Deaths'].rolling(21).sum().shift(-20).reset_index(0,drop=True)
df_states['7ma_death'] = df_states['7sum_death'] / 7
df_states['14ma_death'] = df_states['14sum_death'] / 14
df_states['21ma_death'] = df_states['21sum_death'] / 21
```
